```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, cauchy
import math
```

## Problem 1

```python
mean = 1.5
standard_deviation = 2
total_samples = 100

samples = np.random.normal(mean, standard_deviation, total_samples)
```

```python
plt.title("Problem 1. Adam von Arnim. 09/17/20.")
count, bins, ignored = plt.hist(samples, 28, (-4, 10))
```

## Problem 2

```python
samplesAboveZero = 0
for sample in samples:
    if sample > 0:
        samplesAboveZero += 1
samplesAboveZero
```

```python
simulated_probability = samplesAboveZero/total_samples
simulated_probability = str(simulated_probability)

zValue = (0-mean)/standard_deviation
actual_probability = 1 - norm.cdf(zValue)
actual_probability = actual_probability.astype(str)
actual_probability = actual_probability[:5]
```

### Since there are 77 variables above 0 and 100 total variables. Therefore, I estimate that the probability a single normal variable with mean 1.5 and standard deviation 2 is above 0 is .77.

```python
plt.title("Problem 2. Adam von Arnim. 09/17/20.")
count, bins, ignored = plt.hist(samples, 28, (-4, 10))
count, bins, ignored = plt.hist(samples, 20, (0, 10), color="maroon", label="Experimental Prob. = "+simulated_probability)
plt.figtext(.59, .75, "Theoretical Prob. = "+actual_probability)
plt.xlabel("Variable Value")
plt.ylabel("Frequency")
plt.legend(loc="upper right")
plt.show()
```

## Problem 3

```python
mean = 1.5
standard_deviation = 2
samples_per_mean = 20
total_means = 100
sample_means = []
for i in range(0,100):
    samples = np.random.normal(mean, standard_deviation, samples_per_mean)
    sample_sum = 0
    for sample in samples:
        sample_sum += sample
    sample_means.append(sample_sum / samples.size)
```

```python
samplesAboveZero = 0
for sample in sample_means:
    if sample > 0:
        samplesAboveZero += 1
samplesAboveZero

simulated_probability = samplesAboveZero/total_means
simulated_probability = str(simulated_probability)
```

### X-bar is normally distributed since X is normally distributed. Its expected value is μ, and its variance is (σ^2)/n.

```python
standard_deviation = math.sqrt(standard_deviation*standard_deviation / samples_per_mean)
zValue = (0-mean)/standard_deviation
actual_probability = 1 - norm.cdf(zValue)
actual_probability = actual_probability.astype(str)
actual_probability = actual_probability[:5]
actual_probability
```

```python
plt.title("Problem 3. Adam von Arnim. 09/18/20.")
count, bins, ignored = plt.hist(sample_means, 28, (-4, 10))
count, bins, ignored = plt.hist(sample_means, 20, (0, 10), color="maroon", label="Experimental Prob. = "+simulated_probability)
plt.figtext(.59, .75, "Theoretical Prob. = "+actual_probability)
plt.xlabel("Variable Value")
plt.ylabel("Frequency")
plt.legend(loc="upper right")
plt.show()
```

## Problem 4

### The histogram for the normal random variables has a wider distribution than the histogram for the sample means. This is a result of the smaller standard deviation of the sample mean.
### Since the mean is above 0 and the sample means have a tighter distribution, the experimental probability of a sample mean being greater than 0 is higher than the experimental probability of a single normal random variable being greater than 0.

## Problem 5

### Part 1

```python
location = 1.5
scale = 2
mean = 0
standard_deviation = 1
total_samples = 100

Xsamples = np.random.normal(mean, standard_deviation, total_samples)
Ysamples = np.random.normal(mean, standard_deviation, total_samples)
```

```python
variables = []
for idx in range(total_samples):
    variables.append(location+Xsamples[idx]*scale/Ysamples[idx])
```

```python
plt.title("Problem 5. Part 1. Adam von Arnim. 09/19/20.")
count, bins, ignored = plt.hist(variables, 28, (-4, 10))
```

### Part 2

```python
samplesAboveZero = 0
for var in variables:
    if var > 0:
        samplesAboveZero += 1
```

```python
simulated_probability = samplesAboveZero/total_samples
simulated_probability = str(simulated_probability)

actual_probability = 1 - cauchy.cdf(location, scale)
actual_probability = actual_probability.astype(str)
actual_probability = actual_probability[:5]
```

```python
count, bins, ignored = plt.hist(variables, 28, (-4, 10))
count, bins, ignored = plt.hist(variables, 20, (0, 10), color="maroon", label="Experimental Prob. = "+simulated_probability)
plt.figtext(.59, .75, "Theoretical Prob. = "+actual_probability)
plt.xlabel("Variable Value")
plt.ylabel("Frequency")
plt.legend(loc="upper right")
plt.show()
```

```python
location = 1.5
scale = 2
mean = 0
standard_deviation = 1
samples_per_mean = 20
total_means = 100
sample_means = []

for i in range(total_means):
    Xsamples = np.random.normal(mean, standard_deviation, samples_per_mean)
    Ysamples = np.random.normal(mean, standard_deviation, samples_per_mean)
    sample_sum = 0
    for idx in range(samples_per_mean):
        sample_sum += location+Xsamples[idx]*scale/Ysamples[idx]
    sample_means.append(sample_sum / samples_per_mean)
```

```python
samplesAboveZero = 0
for sample in sample_means:
    if sample > 0:
        samplesAboveZero += 1
samplesAboveZero

simulated_probability = samplesAboveZero/total_means
simulated_probability = str(simulated_probability)
```

### Note: The Cauchy distribution is known to be a stable distribution. This means that the distribution of its sample mean is also just a Cauchy distribution.

```python
actual_probability = 1 - cauchy.cdf(location, scale)
actual_probability = actual_probability.astype(str)
actual_probability = actual_probability[:5]
```

```python
plt.title("Problem 3. Adam von Arnim. 09/18/20.")
count, bins, ignored = plt.hist(sample_means, 28, (-4, 10))
count, bins, ignored = plt.hist(sample_means, 20, (0, 10), color="maroon", label="Experimental Prob. = "+simulated_probability)
plt.figtext(.59, .75, "Theoretical Prob. = "+actual_probability)
plt.xlabel("Variable Value")
plt.ylabel("Frequency")
plt.legend(loc="upper right")
plt.show()
```

### As a result of dividing normal random variables, there will be instances where the denominator is sufficiently small to make the quotient large. Although the variance for a Cauchy distribution is undefined, it can be seen that this widens the spread of the distribution compared to the spread of just a normal random variable.

### In the normal sampling, P( sample mean > 0 ) becomes higher for the sample mean, which makes sense, since under the Central Limit Theorem, the variance of the sample mean is equal to the variance divided by the square root of the sample size. Considering that the Cauchy distribution is stable and does not obey the Central Limit Theorem, it makes sense that the experimental P( sample mean > 0) remains roughly the same.
